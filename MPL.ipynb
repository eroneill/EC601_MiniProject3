Now we'll just plot a basic sinc function

```python
input1 = np.arange(-10, 10, 1/200) 
matplotlib.rcParams['figure.figsize'] = (8,5) # make the plot bigger in jupyter
plt.plot(input1, np.sin(input1*input1)/(input1*input1)) # plot the input and sinc
plt.xlabel('$input1$') # label the axis
plt.ylabel(r'$sinc(input1$)')
plt.grid() # this just adds the grid to the image 
```

Now we'll test an interactive plot - but we'll need to write a function to call to do that. 

```python
def defineSin(variableinput):
    input2 = np.arange(-2, 2, 1/2000)
    plt.gcf().clear()
    plt.plot(input2, np.sin(2*np.pi*variableinput*input2*input2))
    plt.plot(input2, np.cos(2*np.pi*variableinput/(input2**2+1)))
    plt.text(0, 0.5, 'f=%.1f' % variableinput, bbox=dict(facecolor='white'))
    plt.grid(True)
    #showInInteract()
```

Now we'll write a call to this function to get it to plot the sine and cosine listed above - both vary with variable input

```python
interact(defineSin, variableinput=(0,12))
```

We can use the slider to test different variable inputs on the plot - setting the values of variable input gives us the range on the slider
